```python
# Display a single Neuron

from carabao.graphics import Neurons
import matplotlib.pyplot as plt

nr = Neurons(1,1)
#can = nr.canvas()
nr.cell((0,0),x=0,y=1)
nr.input(0,1)
plt.show()
```

```python
# Display a Layer

from carabao.graphics import Neurons
import matplotlib.pyplot as plt

nr = Neurons(4,8)
#can = nr.canvas()
for j in range(0,nr.n):
    nr.cell((0,j),x=0,y=0)
    nr.cell((1,j),x=1,y=0)
    nr.cell((2,j),x=0,y=1)
    nr.cell((3,j),x=1,y=1)
    nr.input(j,1 if j < 3 else 0)

plt.show()
```

```python
# Display a Minicolumn
# with given permanences P and synaptics Q 

from carabao.graphics import Neurons
import matplotlib.pyplot as plt
import torch

    # define permanences

P = torch.tensor([[0.5,0.4,0.3,0,0],[0.3,0,0.5,0,0.5]])
print("P:\n",P)

    # setup synaptics Q

Q = (P >= 0.5).int()
print("Q:\n",Q)

    # visualize

nr = Neurons(4,1)
#can = nr.canvas()
nr.cell((0,0),0,1,P*0.0)
nr.cell((1,0),0,1,P+0.0,Q)
nr.cell((2,0),0,1,P+0.1,Q)
nr.cell((3,0),0,1,P+0.2,Q)
nr.input(0,1)
plt.show()
```
