```python
# Display a single Neuron

from carabao.screen import Screen

scr = Screen('Neurons',1,1)
scr.cell((1,0),x=0,y=1)
scr.input(0,1)
scr.show()
```

```python
# Display a Layer

from carabao.screen import Screen

scr = Screen('Neurons',4,8)
for j in range(0,scr.n):
    scr.cell((1,j),x=0,y=0)
    scr.cell((2,j),x=1,y=0)
    scr.cell((3,j),x=0,y=1)
    scr.cell((4,j),x=1,y=1)
    scr.input(5,1 if j < 3 else 0)

scr.show()
```
