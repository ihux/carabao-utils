```python
# Display a single Neuron

from carabao.screen import Screen

scr = Screen('Neurons',1,1)
scr.cell((0,0),x=0,y=1)
scr.input(0,1)
scr.show()
```

```python
# Display a Layer

from carabao.screen import Screen

scr = Screen('Neurons',4,8)
for j in range(0,scr.n):
    scr.cell((0,j),x=0,y=0)
    scr.cell((1,j),x=1,y=0)
    scr.cell((2,j),x=0,y=1)
    scr.cell((3,j),x=1,y=1)
    scr.input(j,1 if j < 3 else 0)

scr.show()
```

```python
# Display a Minicolumn
# with given permanences P and synaptics Q 

from carabao.screen import Screen
#import matplotlib.pyplot as plt
import torch

    # define permanences

P = torch.tensor([[0.5,0.4,0.3,0,0],[0.3,0,0.5,0,0.5]])
print("P:\n",P)

    # setup synaptics Q

Q = (P >= 0.5).int()
print("Q:\n",Q)

    # visualize

scr = Screen('Neurons',4,1)
scr.cell((0,0),0,1,P*0.0)
scr.cell((1,0),0,1,P+0.0,Q)
scr.cell((2,0),0,1,P+0.1,Q)
scr.cell((3,0),0,1,P+0.2,Q)
scr.input(0,1)
scr.show()
```
