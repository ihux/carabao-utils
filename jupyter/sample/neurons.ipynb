```python
# Display a single Neuron

from carabao.graphics import Neurons
import matplotlib.pyplot as plt

nrs = Neurons(1,1)
can = nrs.canvas()
nrs.cell((0,0),x=0,y=1)
nrs.input(0,1)
plt.show()
```

```python
# Display a Layer

from carabao.graphics import Neurons
import matplotlib.pyplot as plt

nrs = Neurons(4,8)
can = nrs.canvas()
for j in range(0,nrs.n):
    nrs.cell((0,j),x=0,y=0)
    nrs.cell((1,j),x=1,y=0)
    nrs.cell((2,j),x=0,y=1)
    nrs.cell((3,j),x=1,y=1)
    nrs.input(j,1 if j < 3 else 0)

plt.show()
```

```python
# Display a Minicolumn

from carabao.graphics import Neurons
import matplotlib.pyplot as plt
import torch

P = torch.zeros(5,2)
P[0,0] = 0.5;  P[1,0] = 0.4;  P[2,0] = 0.3
P[0,1] = 0.3;  P[2,1] = 0.4;  P[4,1] = 0.5
print("P:\n",P)

nrs = Neurons(4,1)
can = nrs.canvas()
nrs.cell((0,0),0,1,P*0.0)
nrs.cell((1,0),0,1,P+0.0)
nrs.cell((2,0),0,1,P+0.1)
nrs.cell((3,0),0,1,P+0.2)
nrs.input(0,1)
plt.show()
```
